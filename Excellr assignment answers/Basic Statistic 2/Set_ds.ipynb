```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("Book.xlsx")
```

```python
companies = data["Name of company"]
measure_X = data["Measure X"]


plt.figure(figsize=(10, 6))
plt.barh(companies, measure_X, color='skyblue')
plt.xlabel('Measure X (%)')
plt.title('Measure X for Different Companies')
plt.gca().invert_yaxis()  
plt.show()


mean_X = np.mean(measure_X)


std_dev_X = np.std(measure_X)


variance_X = np.var(measure_X)

print("Mean ():", mean_X)
print("Standard Deviation ():", std_dev_X)
print("Variance (^2):", variance_X)

```

```python
import numpy as np


returns = np.array([-2000, -1000, 0, 1000, 2000, 3000])


probabilities = np.array([0.1, 0.1, 0.2, 0.2, 0.3, 0.1])


expected_return = np.sum(returns * probabilities)


standard_deviation = np.sqrt(np.sum((returns - expected_return)**2 * probabilities))

print("Standard deviation of returns:", standard_deviation)
```
