```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 250
import warnings
warnings.filterwarnings('ignore')
sns.set_theme(style='darkgrid', palette='viridis')
%matplotlib inline
```

```python
df = pd.read_csv("book.csv", encoding='latin-1')
df = df.drop(columns='Unnamed: 0')
df
```

```python
df['User.ID'].nunique()
```

```python
df['Book.Title'].nunique()
```

```python
df = df.drop_duplicates(subset=['Book.Title'])
```

```python
pivot = pd.pivot(data=df, index ='Book.Title',columns='User.ID', values='Book.Rating')
```

```python
pivot
```

```python
pivot.fillna(0,inplace=True)
```

```python
pivot
```

```python
from sklearn.metrics import pairwise_distances
```

```python
sim = pd.DataFrame(1 - pairwise_distances(pivot, metric='cosine'))
sim = pd.DataFrame(sim)
```

```python
sim_array = sim.values
```

```python
np.fill_diagonal(sim_array,0)
```

```python
similarity = pd.DataFrame(sim)
similarity
```

```python
similarity.index = df['Book.Title'].unique()
similarity.columns = df['Book.Title'].unique()
```
