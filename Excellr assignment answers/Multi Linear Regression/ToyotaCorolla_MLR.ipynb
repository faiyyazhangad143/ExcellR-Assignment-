# Cook's Distance
   - **Detecting influencers and outliers**

```python
model
```

```python
#Find the influence data
inf = model.get_influence()

#Calculate the cooks distance
c,p = inf.cooks_distance
```

```python
c
```

```python
# Cook's distance plot
plt.stem(c)
plt.xlabel('Index')
plt.ylabel('Cooks Distance')
```

```python
np.argmax(c), np.max(c)
```
